```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv("Subquery Results.csv")
```

```python
#When did a student with ID 268727 first watch a lecture?

df[df['student_id']==268727]
```

```python
no_of_purchases = df['first_date_purchased'].notna().sum()
no_of_watches = df['first_date_watched'].notna().sum()
```

```python
no_of_purchases,no_of_watches
```

```python
conversion_rate = round(no_of_purchases/no_of_watches,2)*100
av_reg_watch = round(df['days_diff_reg_watch'].mean(),2)
av_watch_purch = round(df['days_diff_watch_purch'].mean(),2)
```

```python
conversion_rate,av_reg_watch,av_watch_purch
```

```python
median_reg_watch = df['days_diff_reg_watch'].median()
mode_reg_watch = df['days_diff_reg_watch'].mode()
```

```python
median_reg_watch
```

```python
av_reg_watch,median_reg_watch,mode_reg_watch
```

```python
df['days_diff_reg_watch'].dropna().plot(kind='hist', bins=30, edgecolor='black')

plt.title('Distribution of Days Difference between Registration and First Watch')
plt.xlabel('Days Difference')
plt.ylabel('Frequency')
```

```python
median_watch_purch = df['days_diff_watch_purch'].median()
mode_watch_purch = df['days_diff_watch_purch'].mode()
```

```python
av_watch_purch,median_watch_purch,mode_watch_purch
```

```python
df['days_diff_watch_purch'].dropna().plot(kind='hist', bins=30, edgecolor='black')

plt.title('Distribution of Days Difference between First Watch and Purchase')
plt.xlabel('Days Difference')
plt.ylabel('Frequency')
```
